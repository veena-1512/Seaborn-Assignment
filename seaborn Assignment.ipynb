Q 1.  Seaborn is a popular data visualization library built on top of Matplotlib in Python. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five common plots that can be created using Seaborn and their typical uses:

i. Scatter Plot:

Use: Scatter plots are used to visualize the relationship between two numerical variables. Each data point is represented as a point on the plot, where the x-axis represents one variable, and the y-axis represents the other variable.

ii. Bar Plot:
Use: Bar plots are used to display and compare categorical data. They represent data using rectangular bars, with the length of each bar corresponding to the quantity or frequency of the data category it represents.

iii. Histogram:

Use: Histograms are used to visualize the distribution of a single numerical variable. It divides the data into intervals (bins) and shows the frequency or count of data points that fall into each bin.

iv. Box Plot (Box-and-Whisker Plot):
Use: Box plots are used to display the distribution of numerical data and identify potential outliers. They show the median, quartiles, and potential outliers of the data, providing a good summary of its statistical properties.

v. Heatmap:
Use: Heatmaps are used to visualize the correlation between variables in a dataset. They display a grid where each cell is colored based on the relationship between two variables, making it easy to identify patterns and relationships.

Q 2. To plot a line plot using the "fmri" dataset in Seaborn, you first need to import the necessary libraries and load the dataset using the load_dataset function. Then, you can use the lineplot function in Seaborn to create the line plot with "timepoint" on the x-axis and "signal" on the y-axis, grouping the data by different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot a line plot with x="timepoint" and y="signal", grouped by "event" and "region"
sns.set(style="darkgrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

plt.title("FMRI Signal by Timepoint for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

Q 3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")

# Plot the first box plot with x='pclass' and y='age'
plt.figure(figsize=(8, 6))
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title("Age Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

plt.show()

```

```python
# Plot the second box plot with x='pclass' and y='fare'
plt.figure(figsize=(8, 6))
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

plt.show()
```

Q. 4.o plot a histogram for the 'price' column in the "diamonds" dataset from Seaborn, and use the hue parameter to differentiate the data based on the 'cut' column, you can follow the steps below:

```python

diamonds_data = sns.load_dataset("diamonds")


plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', bins=30, kde=True, palette='Set1')

plt.title("Price Distribution by Cut for Diamonds")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```

Q. 5.  Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset. 

```python
iris_data = sns.load_dataset("iris")

sns.pairplot(iris_data, hue='species', palette='Set1')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights_data = sns.load_dataset("flights")

flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d", cbar_kws={'label': 'Number of Passengers'})
plt.title("Number of Passengers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")

plt.show()
```
